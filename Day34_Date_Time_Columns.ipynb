* By default date and time columns are read as strings by pandas __CHANGE IT__ to date and time by pd.to_datetime

```python
import numpy as np
import pandas as pd
```

```python
date=pd.read_csv("orders.csv")
time=pd.read_csv("messages.csv")
```

```python
date.info() #see below date column is object
```

### 1. Working With Dates

```python
date['date']=pd.to_datetime(date['date'])
date.info()
```

### 2. Extract Year

```python
date['date_year'] = date['date'].dt.year
# column add hrha ha bhai abhi tk tujhy smjh ni ayi? df['date_year]=.....
```

### 3. Extract Month

```python
date["date_month_name"]=date['date'].dt.month_name()
```

### 4. Extract Day

```python
date['date_day'] = date['date'].dt.day
```
